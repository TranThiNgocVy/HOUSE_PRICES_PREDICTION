## 🎯 Task 1 — Tại Sao Cần Dự Đoán Giá Nhà?

---

### 🏚️ Bài Toán Định Giá Bất Động Sản

Giá nhà phụ thuộc vào **hàng chục yếu tố** khác nhau — từ diện tích, vị trí, số phòng ngủ, đến chất lượng vật liệu xây dựng và tình trạng khu phố xung quanh.

Việc định giá chính xác là **nhu cầu thiết yếu** đối với nhiều bên liên quan trong thị trường bất động sản:

| 👤 Đối tượng | 📌 Nhu cầu | 💬 Ví dụ thực tế |
|---|---|---|
| 🏠 **Người mua** | Biết giá hợp lý để không mua đắt | Nhà 80m², 3 phòng ngủ → giá bao nhiêu? |
| 💰 **Người bán** | Định giá chính xác để bán nhanh | Đặt giá quá cao → không ai mua |
| 🏦 **Ngân hàng** | Thẩm định giá trước khi cho vay | Cho vay theo % giá trị bất động sản |
| 📊 **Nhà đầu tư** | Tìm cơ hội đầu tư sinh lời | Nhà dưới giá thị trường → mua vào |

---

### 🤔 Vấn Đề Với Cách Định Giá Truyền Thống

Trước khi có ML, việc định giá nhà thường phụ thuộc vào:

- 👨‍💼 **Môi giới / chuyên gia:** Kinh nghiệm cá nhân, dễ thiên lệch
- 📋 **So sánh thủ công:** Tìm nhà tương đồng trong khu vực, tốn thời gian
- 📉 **Công thức đơn giản:** Giá/m² × diện tích — bỏ qua hàng chục yếu tố khác

**Hạn chế:** Không nhất quán, chậm, và không mở rộng được quy mô.

## 🤖 Task 3 — Phương Pháp: Từ Linear Đến Ensemble

---

### 📈 Tại Sao Cần Nhiều Loại Mô Hình?

Không có một mô hình nào "tốt nhất" cho mọi bài toán. Trong dự án này, chúng ta sẽ thử nghiệm **nhiều loại mô hình khác nhau** và quan sát:

- Mô hình nào học được pattern từ dữ liệu tốt nhất?
- Mô hình nào bị **overfitting** (học thuộc dữ liệu train nhưng kém trên test)?
- Kết hợp nhiều mô hình có cho kết quả tốt hơn không?

---

### 🗺️ Lộ Trình Mô Hình — Từ Đơn Giản Đến Mạnh Mẽ

```
🟢 Bước 1: Linear Regression (Baseline)
    ↓
🟡 Bước 2: Ridge / Lasso / ElasticNet (Regularization)
    ↓
🔴 Bước 3: Random Forest / XGBoost / LightGBM (Tree-based)
    ↓
🏆 Bước 4: Stacking / Blending (Ensemble)
```

---

### 📊 Bảng So Sánh Các Mô Hình

| 🎚️ Giai đoạn | 🤖 Mô hình | 💡 Đặc điểm | 📖 Khi nào dùng? |
|---|---|---|---|
| 🟢 **Cơ bản** | Linear Regression | Đơn giản, dễ giải thích | Baseline đầu tiên |
| 🟡 **Regularization** | Ridge (L2) | Giảm overfitting bằng phạt hệ số lớn | Nhiều features |
| 🟡 **Regularization** | Lasso (L1) | Tự động loại bỏ features không quan trọng | Feature selection |
| 🟡 **Regularization** | ElasticNet | Kết hợp Ridge + Lasso | Cân bằng 2 loại trên |
| 🔴 **Tree-based** | Random Forest | Ensemble cây quyết định, ổn định | Non-linear patterns |
| 🔴 **Tree-based** | XGBoost | Gradient boosting mạnh mẽ, phổ biến Kaggle | Hiệu suất cao |
| 🔴 **Tree-based** | LightGBM | Nhanh hơn XGBoost, tốt với dữ liệu lớn | Dữ liệu nhiều features |
| 🏆 **Ensemble** | Stacking/Blending | Kết hợp dự đoán của nhiều model | Kết quả tốt nhất |

---

### 🎯 Mục Tiêu Đạt Được

| 📏 Metric | 🎯 Mục tiêu | 📌 Ý nghĩa |
|---|---|---|
| 📊 **RMSE (log)** | < 0.12 | Sai số dự đoán trên log scale |
| 📈 **R²** | > 0.90 | Mô hình giải thích được 90%+ biến thiên giá |
| 🏆 **Kaggle Rank** | Top 25% leaderboard | Đủ năng lực thi đấu thực tế |

## 🏆 Task 4 — Mô Tả Sản Phẩm Cuối Cùng

---

### 🎁 Bạn Sẽ Tạo Ra Gì Sau Khóa Học?

Sau khi hoàn thành **9 buổi học**, bạn sẽ có trong tay một **pipeline ML hoàn chỉnh** có thể tái sử dụng cho nhiều bài toán regression khác nhau.

---

### ✨ Các Tính Năng Của Sản Phẩm

| 🔧 Tính năng | 📝 Mô tả |
|---|---|
| 📊 **EDA chuyên sâu** | Phân tích 80+ features, phát hiện outliers, vẽ biểu đồ phân phối & correlation |
| 🧹 **Data Cleaning** | Xử lý missing values thông minh theo từng loại feature |
| 🔧 **Feature Engineering** | Tạo features mới, encoding biến phân loại, log-transformation |
| 🤖 **Nhiều mô hình** | So sánh 5+ thuật toán regression có cross-validation |
| 📈 **Evaluation đầy đủ** | RMSE, MAE, R², biểu đồ so sánh actual vs predicted |
| 🏆 **Kaggle Submission** | File CSV chuẩn để nộp lên Kaggle và xem ranking |

---

### 🏗️ Kiến Trúc Pipeline Tổng Thể

```
📥 Kaggle Data
      ↓
📊 EDA (Exploratory Data Analysis)
   → Phân tích phân phối, correlation, outliers
      ↓
🧹 Data Cleaning
   → Xử lý missing values, sửa kiểu dữ liệu
      ↓
🔧 Feature Engineering
   → Tạo features mới, encoding, scaling
      ↓
🤖 Model Training
   → Linear, Ridge, XGBoost, LightGBM
      ↓
📈 Evaluation & Tuning
   → Cross-validation, hyperparameter tuning
      ↓
🔗 Ensemble (Stacking/Blending)
   → Kết hợp nhiều model tốt nhất
      ↓
📤 Kaggle Submission
   → submission.csv → upload lên Kaggle
```

---

### 🛠️ Công Nghệ Sử Dụng

| 🧩 Thành phần | 🛠️ Công nghệ | 🎯 Vai trò |
|---|---|---|
| 📋 **Dataset** | Kaggle API / Download thủ công | Tải và load dữ liệu |
| 📊 **EDA** | Matplotlib, Seaborn | Visualization phân phối, correlation |
| 🧹 **Cleaning** | Pandas, NumPy | Missing values, outliers |
| 🔧 **Features** | Scikit-learn | Encoding, scaling, feature creation |
| 🤖 **Models** | Scikit-learn, XGBoost, LightGBM | Training và hyperparameter tuning |
| 📈 **Evaluation** | Scikit-learn cross_val_score | RMSE, R² trên nhiều folds |

---

> 🚀 **Quan trọng:** Pipeline bạn xây dựng không chỉ dùng được cho bài toán này — bạn có thể **tái sử dụng toàn bộ** cấu trúc cho các bài toán regression khác như: dự đoán lương, dự đoán doanh thu, hay dữ liệu BDS Việt Nam!
