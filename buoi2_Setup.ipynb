### 🏆 KPI (Key Performance Indicators) — Mục Tiêu Kaggle

Đây là **cam kết chất lượng** của dự án này. Sau khi hoàn thành, mô hình của bạn cần đạt:

```
📊 Hiệu năng mô hình:
   ├── RMSE (log scale):  < 0.12   ← Sai số dự đoán thấp
   ├── R² Score:          > 0.90   ← Mô hình giải thích 90%+ biến thiên
   └── Kaggle Rank:       Top 25%  ← Cạnh tranh thực sự

🎯 Giá trị thực tiễn:
   ├── Dự đoán giá nhà với độ lệch ±10%
   └── Xác định 10 yếu tố quan trọng nhất ảnh hưởng đến giá
```

#### 📏 RMSE Là Gì? Tại Sao Quan Trọng?

**RMSE (Root Mean Squared Error)** — Sai số căn bậc hai trung bình:

$$RMSE = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2}$$

Trong dự án này, chúng ta dùng **RMSE trên log scale** (tức là tính RMSE sau khi đã log transform SalePrice):

- **RMSE = 0.12** nghĩa là dự đoán sai khoảng **±12%** giá trị log → tương đương ~**12-15% giá thực tế**
- Ví dụ: Nhà thực giá **$200,000** → mô hình dự đoán trong khoảng **$170,000 – $230,000**

#### 📈 R² Score Là Gì?

**R² (R-squared / Hệ số xác định)**:

$$R^2 = 1 - \frac{\sum(y_i - \hat{y}_i)^2}{\sum(y_i - \bar{y})^2}$$

- **R² = 1.0** → mô hình dự đoán **hoàn hảo 100%**
- **R² = 0.90** → mô hình giải thích được **90% sự biến thiên** của giá nhà
- **R² = 0.0** → mô hình không tốt hơn việc đoán giá trung bình

> 💡 **Ghi nhớ:** R² = 0.90 là mức rất tốt cho bài toán định giá bất động sản thực tế!

### 📂 Load Dataset — Đọc Dữ Liệu Lần Đầu

---

#### 🤔 `pd.read_csv()` Hoạt Động Như Thế Nào?

File CSV (Comma-Separated Values) là định dạng lưu dữ liệu bảng phổ biến nhất:

```
Id,MSSubClass,LotFrontage,LotArea,SalePrice
1,60,65,8450,208500
2,20,80,9600,181500
3,60,68,11250,223500
```

Khi dùng `pd.read_csv('train.csv')`, Pandas sẽ:
1. Mở file CSV
2. Đọc dòng đầu tiên → đặt làm tên cột (header)
3. Đọc các dòng còn lại → lưu vào DataFrame (bảng 2 chiều trong bộ nhớ)
4. Tự động phát hiện kiểu dữ liệu (số, chữ, ngày...)

---

#### 📏 `DataFrame.shape` Là Gì?

Mỗi DataFrame có thuộc tính `.shape` trả về tuple `(số_hàng, số_cột)`:

```python
train_df.shape  → (1460, 81)
#                   ↑      ↑
#               1460 nhà  81 cột (80 features + 1 target SalePrice)
```

- **Train có 81 cột** = 80 features + cột `SalePrice` (target)
- **Test có 80 cột** = 80 features (không có `SalePrice` — đó là thứ ta cần dự đoán!)

#### 🔍 Các Lệnh Khám Phá Dữ Liệu Cơ Bản

| Lệnh | Tác dụng | Khi nào dùng |

| `df.head(5)` | Xem 5 hàng đầu | Nhìn nhanh dữ liệu trông như thế nào |

> 💡 **Quy tắc vàng:** Luôn chạy `df.info()` và `df.describe()` ngay sau khi load dữ liệu mới — đây là bước **"gặp gỡ dataset lần đầu"** không thể bỏ qua!

## 📊 Task 3 — Phân Tích Biến Mục Tiêu `SalePrice`

---

### 🤔 Tại Sao Phải Phân Tích `SalePrice` Trước?

`SalePrice` là **thứ chúng ta cần dự đoán** — hiểu rõ phân phối của nó là **bước bắt buộc** trước khi train bất kỳ mô hình nào.

Hầu hết các thuật toán regression (đặc biệt là Linear Regression) đều có giả định:
> **Giả định:** Biến target phải xấp xỉ phân phối **chuẩn (Normal Distribution)** — tức là hình chuông đối xứng

Nếu target bị **skewed (lệch)** mà không xử lý → mô hình học kém, kết quả tệ!

---

### 📏 Skewness (Độ Lệch) Là Gì?

**Skewness** đo mức độ "lệch" của phân phối so với đường chuẩn:

```
Skewness = 0       → Phân phối hoàn toàn đối xứng (lý tưởng)
Skewness > 0       → Lệch phải (right-skewed): đuôi dài về bên phải
Skewness < 0       → Lệch trái (left-skewed): đuôi dài về bên trái

Quy tắc thực tế:
  |skewness| < 0.5  → Gần như bình thường ✅
  |skewness| 0.5–1  → Lệch vừa ⚠️
  |skewness| > 1    → Lệch nhiều ❌ → Cần xử lý!
```

**SalePrice có skewness ≈ 1.88** → Lệch phải nghiêm trọng!

Ý nghĩa thực tế: Có **rất nhiều nhà giá trung bình** ($150k–$250k) và **một ít nhà siêu đắt** ($500k+) — những nhà đắt này kéo đuôi phân phối về bên phải.

---

### 🔧 Giải Pháp: Log Transformation

**`np.log1p(x)`** = log(x + 1) — biến đổi logarithm:

```
Tại sao dùng log1p thay vì log?
  → log(0) = undefined (lỗi toán học!)
  → log1p(0) = log(0+1) = log(1) = 0 ✅ (an toàn với giá trị 0)
```

Sau khi log transform:
- Skewness giảm từ **1.88 → ~0.12** (gần normal) ✅
- Phân phối trở nên đối xứng hơn
- Mô hình Linear Regression học tốt hơn nhiều

**⚠️ Quan trọng:** Khi predict xong, phải **đảo ngược** transform:
```python
y_pred_original = np.expm1(y_pred_log)  # expm1(x) = e^x - 1
```

---

### 📈 Q-Q Plot Là Gì?

**Q-Q Plot (Quantile-Quantile Plot)** — biểu đồ so sánh phân phối thực tế với phân phối chuẩn:
- **Các điểm nằm trên đường thẳng chéo** → phân phối gần chuẩn ✅
- **Điểm cong lên/xuống ở đuôi** → phân phối bị skewed ❌

> 💡 Sau khi log transform, Q-Q plot sẽ thẳng hơn nhiều — đó là dấu hiệu tốt!

````python
# ============================================================
# 📊 PHÂN TÍCH PHÂN PHỐI SalePrice — Original vs Log
# ============================================================

fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('📊 Phân Phối SalePrice: Original vs Log-Transformed', 
             fontsize=14, fontweight='bold', y=1.02)

# --- Biểu đồ 1: Phân phối gốc (Original) ---
axes[0].hist(train_df['SalePrice'], bins=50, color='steelblue', 
             edgecolor='white', alpha=0.8)
axes[0].axvline(train_df['SalePrice'].mean(),   color='red',    linestyle='--', 
                linewidth=2, label=f"Mean: ${train_df['SalePrice'].mean():,.0f}")
axes[0].axvline(train_df['SalePrice'].median(), color='orange', linestyle='--', 
                linewidth=2, label=f"Median: ${train_df['SalePrice'].median():,.0f}")
axes[0].set_xlabel('SalePrice (USD)', fontsize=11)
axes[0].set_ylabel('Frequency', fontsize=11)
axes[0].set_title('❌ Original — Lệch phải', fontsize=12)
axes[0].legend(fontsize=9)
axes[0].tick_params(axis='x', rotation=30)

# --- Biểu đồ 2: Sau log transform ---
log_price = np.log1p(train_df['SalePrice'])
axes[1].hist(log_price, bins=50, color='coral', 
             edgecolor='white', alpha=0.8)
axes[1].axvline(log_price.mean(),   color='red',    linestyle='--', 
                linewidth=2, label=f"Mean: {log_price.mean():.3f}")
axes[1].axvline(log_price.median(), color='orange', linestyle='--', 
                linewidth=2, label=f"Median: {log_price.median():.3f}")
axes[1].set_xlabel('log1p(SalePrice)', fontsize=11)
axes[1].set_ylabel('Frequency', fontsize=11)
axes[1].set_title('✅ Log-Transformed — Gần chuẩn', fontsize=12)
axes[1].legend(fontsize=9)

# --- Biểu đồ 3: Q-Q Plot ---
stats.probplot(log_price, plot=axes[2])
axes[2].set_title('📈 Q-Q Plot (Log-Transformed)', fontsize=12)
axes[2].get_lines()[1].set_color('red')  # Đường tham chiếu màu đỏ

plt.tight_layout()
plt.show()

# --- Kết quả Skewness ---
print("📊 Phân tích Skewness:")
print(f"   Original SalePrice : {train_df['SalePrice'].skew():.3f}  ← Lệch phải nghiêm trọng!")
print(f"   Log-Transformed    : {log_price.skew():.3f}  ← Gần chuẩn ✅")
print(f"\n💡 Kết luận: Log transform giảm skewness từ {train_df['SalePrice'].skew():.2f} → {log_price.skew():.2f}")
````

| Hàm | Công thức | Lưu ý |
|---|---|---|

````python
# ============================================================
# 🔧 ÁP DỤNG LOG TRANSFORM — Chuẩn bị biến target cho modeling
# ============================================================

# --- Áp dụng log1p transform ---
y = np.log1p(train_df['SalePrice'])

# --- Thông tin trước và sau transform ---
print("🔧 Log Transformation Applied!")
print("\n📊 So sánh trước và sau transform:")
print(f"{'Thống kê':<18} {'Original':>15} {'Log-Transformed':>18}")
print("-" * 55)
print(f"{'Min':<18} {train_df['SalePrice'].min():>15,.0f} {y.min():>18.4f}")
print(f"{'Max':<18} {train_df['SalePrice'].max():>15,.0f} {y.max():>18.4f}")
print(f"{'Mean':<18} {train_df['SalePrice'].mean():>15,.2f} {y.mean():>18.4f}")
print(f"{'Std Dev':<18} {train_df['SalePrice'].std():>15,.2f} {y.std():>18.4f}")
print(f"{'Skewness':<18} {train_df['SalePrice'].skew():>15.3f} {y.skew():>18.3f}")
print()
print("✅ Biến y (target) đã sẵn sàng để dùng trong modeling!")

# --- Kiểm tra: inverse transform ---
print("\n🔄 Kiểm tra inverse transform:")
sample_log  = y.iloc[0]
sample_orig = np.expm1(sample_log)
print(f"   Log value          : {sample_log:.4f}")
print(f"   After expm1()      : ${sample_orig:,.0f}")
print(f"   Actual SalePrice   : ${train_df['SalePrice'].iloc[0]:,.0f}")
print(f"   Match: {'✅ Khớp!' if abs(sample_orig - train_df['SalePrice'].iloc[0]) < 1 else '❌ Sai!'}")
````
